## Q16:   Create a horizontal frequency barplot for 'region'

````python
num_cols = df['region'].unique()
bar_widths = df['region'].value_counts()
bar_positions = np.arange(3) + 0.75
tick_positions = range(1,4)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.barh(bar_positions, bar_widths, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_ylabel("Region")
ax.set_xlabel("Frequency")
plt.show()
````

## Q17:   Create a horizontal frequency barplot for 'sub_region'

````python
num_cols = df['sub_region'].unique()
bar_widths = df['sub_region'].value_counts()
bar_positions = np.arange(9) + 0.75
tick_positions = range(1,10)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.barh(bar_positions, bar_widths, 0.5)
ax.set_yticks(tick_positions)
ax.set_yticklabels(num_cols)
ax.set_ylabel("Sub Region")
ax.set_xlabel("Frequency")
plt.show()
````
